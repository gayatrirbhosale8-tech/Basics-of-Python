# Student Marks Management System



```python
Students_Marks = [39,35.5,39.9,37,36,]
print("average =", 39 + 35.5+ 39.9 +37 +36 /5)
print('Max=',max(Students_Marks))
print('Sum=',sum(Students_Marks))
print('Min=',min(Students_Marks))
print('Length=',len(Students_Marks))
sort_Students_Marks=sorted(Students_Marks)
print('ascending order=',sort_Students_Marks)
```

```python
'''2. grocery shopping list
a user maintains a grocery list for monthly shopping
Tasks:
 Create a grocery list
 Add new items to the list
 Remove purchased items
 Display the final shopping list
 Concept used: append(), remove(), indexing'''
```

```python
GroceryList = ["Milk","Bread","Cheese","Orange","Wheat"]
print('grocery list=',GroceryList)
GroceryList.insert(4,"Rice")
print('new items to the list=',GroceryList)
Daily = ["Soap","Oil","Biscuits"]
GroceryList.extend(Daily)
print('new items to the list=',GroceryList)
GroceryList.remove("Orange")
print('Remove purchased items=',GroceryList)
print('final shopping list=',GroceryList)
```

```python
'''3. A person stores phone contacts in a list.
Task:
Create a list of contact
Add a new contact
Delete an existing contact
Display contacts alphabetically
Concept Used: append(),remove(),sort()'''
```

```python
ContactList = ["Dhvan","Neha","Atharva","Bedang","Tannishtha"]
print("list of contact=",ContactList)
ContactList.append("Gayatri")
print('Add a new contact=',ContactList)
ContactList.remove("Gayatri")
print('Delete an existing contact',ContactList)
ContactList.sort()
print('Display contacts alphabetically=',ContactList)
```

```python
'''4.Tempreture analysis
Daily temperature reading of a city are sorted in a list for 10 days.
Tasks:
Display temperture of first and last 5 days
Find highest and lowest temperature
Calculate average temperature
Concept used: Slicing,max(),min(),sum()'''
```

```python
Temperature = [21,22,18,19,22,14,15,27,24,20]
print("average =", sum(Temperature)/len(Temperature))
print("Max Temperature is =" ,max(Temperature))
print(sum(Temperature))
print('Min Temperature is =',(Temperature))
print('First 5 days is = ',Temperature[0:5])
print('Next 5 days is = ',Temperature[-5:])
```
