```python
#Simple Calculator
a = int(input("Enter 1st Number:"))
b = int(input("Enter 2nd Number:"))
operation = input("Enter the operation: ")
if operation == "+":
  print(a+b)
if operation == "-":
  print(a-b)
if operation == "*":
  print(a*b)
if operation == "/":
  print(a/b)
```
