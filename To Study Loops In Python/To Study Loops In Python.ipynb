```python
s = input ("Enter a string: ")
i,j=0,len(s)-1
palindrome = True
while i<j:
  if s[i]!=s[j]:
    palindrome = False
    break
  i+=1
  j -=1
if palindrome:
  print("String Is A Palindrome")
else:
  print("String Is Not A Palindrome")
```
